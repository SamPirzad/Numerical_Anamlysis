## Question 1

Plot each one of these functions with their error in an appropriate interval:



A) $sin(x) , x$

B) $sin(x) , x-(x^3/3!)$

C) $sin(x) , x-(x^3/3!)+(x^5/5!)-(x^7/7!)$

D) $cos(x) , 1$

E) $cos(x) , 1-(x^2/2!)$

F) $cos(x) , 1-(x^2/2!)+(x^4/4!)$

G) $cos(x) , 1-(x^2/2!)+(x^4/4!)-(x^6/6!)$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Set the interval
x = np.linspace(-np.pi, np.pi, 100)
```

```python

# Set the functions
sin_x = np.sin(x)
cos_x = np.cos(x)
```

```python

# Set the series
series_b = x - (x**3)/6
series_c = x - (x**3)/6 + (x**5)/120 - (x**7)/5040
series_e = 1 - (x**2)/2
series_f = 1 - (x**2)/2 + (x**4)/24
series_g = 1 - (x**2)/2 + (x**4)/24 - (x**6)/720
```

```python

# Set the errors
error_b = np.abs(sin_x - series_b)
error_c = np.abs(sin_x - series_c)
error_e = np.abs(cos_x - series_e)
error_f = np.abs(cos_x - series_f)
error_g = np.abs(cos_x - series_g)
```

```python
# Plot the function A and their error
plt.figure(figsize=(12, 8))
plt.plot(x, sin_x, 'b-', label='sin(x)')
plt.plot(x, x, 'r--', label='x')
plt.legend()
plt.title('a) sin(x), x')
```

```python
# Plot the function B and their error
plt.plot(x, sin_x, 'b-', label='sin(x)')
plt.plot(x, series_b, 'r--', label='x - x^3/3!')
plt.plot(x, error_b, 'g-', label='error')
plt.legend()
plt.title('b) sin(x), x - x^3/3!')
```

```python
# Plot the function C and their error
plt.plot(x, sin_x, 'b-', label='sin(x)')
plt.plot(x, series_c, 'r--', label='x - x^3/3! + x^5/5! - x^7/7!')
plt.plot(x, error_c, 'g-', label='error')
plt.legend()
plt.title('c) sin(x), x - x^3/3! + x^5/5! - x^7/7!')
```

```python
# Plot the function D and their error
plt.plot(x, cos_x, 'b-', label='cos(x)')
plt.plot(x, np.ones(len(x)), 'r--', label='1')
plt.legend()
plt.title('d) cos(x), 1')
```

```python
# Plot the function E and their error
plt.plot(x, cos_x, 'b-', label='cos(x)')
plt.plot(x, series_e, 'r--', label='1 - x^2/2!')
plt.plot(x, error_e, 'g-', label='error')
plt.legend()
plt.title('e) cos(x), 1 - x^2/2!')
```

```python
# Plot the function F and their error
plt.plot(x, cos_x, 'b-', label='cos(x)')
plt.plot(x, series_f, 'r--', label='1 - x^2/2! + x^4/4!')
plt.plot(x, error_f, 'g-', label='error')
plt.legend()
plt.title('f) cos(x), 1 - x^2/2! + x^4/4!')
```

## Question 2

Provide a polynomial based on Taylor series to interpolate function $f(x) = cos(x)$. Your suggested polynomial should has error lower than $10^7$
.

One possible polynomial based on Taylor series to interpolate function $f(x) = cos(x)$ with an error lower than $10^-7$ is:

$cos(x)  1 - (x^2)/2! + (x^4)/4! - (x^6)/6! + (x^8)/8!$

```python
import math
```

```python

def cos_taylor(x):
    # Calculate the first five terms of the Taylor series
    term_2 = -(x ** 2) / math.factorial(2)
    term_4 = (x ** 4) / math.factorial(4)
    term_6 = -(x ** 6) / math.factorial(6)
    term_8 = (x ** 8) / math.factorial(8)
    
    # Sum the terms to obtain the polynomial approximation
    return 1 + term_2 + term_4 + term_6 + term_8
```

```python
# Test the function for different values of x
for x in [-1, 0, 1]:
    approx = cos_taylor(x)
    true = math.cos(x)
    error = abs(approx - true)
    print(f"cos({x}) = {true:.8f}, approx = {approx:.8f}, error = {error:.8f}")
```
