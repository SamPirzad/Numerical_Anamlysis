## Plot Task
Plot periodic function $f(x) = x^3$ in $âˆ’2 < x < 2$ for $n = 2, 5, 10$ with Fourier series. Consider constants $L = 2(2L = 4)$.
Hint: You should approximate given function $f(x)$ with $5, 11$ and $21$ nomials.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Define the function

```python

def f(x):
    return x**3
```

Define the range of x values

```python

x = np.linspace(-2, 2, 1000)
```

Define the number of terms in the Fourier series

```python

N = [5, 11, 21]
```

Set the value of L

```python

L = 2
```

Compute the Fourier series for each value of N

```python

for n in N:
    a0 = (1/L) * np.trapz(f(x), x)
    an = np.zeros(n)
    bn = np.zeros(n)
    for i in range(n):
        an[i] = (1/L) * np.trapz(f(x) * np.cos((i+1)*np.pi*x/L), x)
        bn[i] = (1/L) * np.trapz(f(x) * np.sin((i+1)*np.pi*x/L), x)
    # Compute the Fourier series
    fs = np.zeros(len(x))
    fs += a0/2
    for i in range(n):
        fs += an[i] * np.cos((i+1)*np.pi*x/L) + bn[i] * np.sin((i+1)*np.pi*x/L)
    # Plot the results
    plt.plot(x, fs, label='N={}'.format(n))

# Plot the original function
plt.plot(x, f(x), label='f(x)')

# Set the plot properties
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fourier Series Approximation of f(x) = x^3')
plt.show()
```
